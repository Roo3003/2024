```python
word_search = input.split('\n')

# Extract shape

# n = number of rows
# m = number of cols

n, m = len(word_search), len(word_search[0])

# Going into part 1 expecting part 2 to include wrap-around XMASs... will include functionality for it in this one

def count_XMAS(word_search,row,col, print_finds=False, include_wraps=False):
    '''
    Returns count of XMAS words originating from wordsearch[row][col]
    print_finds toggles printing starting locations of XMAS finds
    include_wraps toggles including wrap-around XMASs
    '''
    xmas_count = 0
    # Just need to check for MAS since we assume we only call this function on an X

    # Going right
    right_1 = word_search[(row) % n][(col+1) % m]
    right_2 = word_search[(row) % n][(col+2) % m]
    right_3 = word_search[(row) % n][(col+3) % m]

    if right_1 == 'M' and right_2 == 'A' and right_3 == 'S':
        
        if include_wraps or (col+3) < m:
            if print_finds:
                print(f'right at {row}, {col}')
            xmas_count += 1
    
    # Going left
    left_1 = word_search[(row) % n][(col-1) % m]
    left_2 = word_search[(row) % n][(col-2) % m]
    left_3 = word_search[(row) % n][(col-3) % m]

    if left_1 == 'M' and left_2 == 'A' and left_3 == 'S':

        if include_wraps or (col-3) >= 0:
            if print_finds:
                print(f'left at {row}, {col}')
            xmas_count += 1
    
    # Going up
    up_1 = word_search[(row-1) % n][(col) % m]
    up_2 = word_search[(row-2) % n][(col) % m]
    up_3 = word_search[(row-3) % n][(col) % m]

    if up_1 == 'M' and up_2 == 'A' and up_3 == 'S':

        if include_wraps or (row-3) >= 0:
            if print_finds:
                print(f'up at {row}, {col}')
            xmas_count += 1
    
    # Going down
    down_1 = word_search[(row+1) % n][(col) % m]
    down_2 = word_search[(row+2) % n][(col) % m]
    down_3 = word_search[(row+3) % n][(col) % m]

    if down_1 == 'M' and down_2 == 'A' and down_3 == 'S':

        if include_wraps or (row+3) < n:
            if print_finds:
                print(f'down at {row}, {col}')
            xmas_count += 1

    # Diagonals

    # -1 => backwards (up or left)
    #  1 => forwards (down or right)

    for vertical_direction in [-1,1]:
        for horizontal_direction in [-1,1]:
            diag_1 = word_search[(row + 1*vertical_direction) % n][(col + 1*horizontal_direction) % m]
            diag_2 = word_search[(row + 2*vertical_direction) % n][(col + 2*horizontal_direction) % m]
            diag_3 = word_search[(row + 3*vertical_direction) % n][(col + 3*horizontal_direction) % m]

            if diag_1 == 'M' and diag_2 == 'A' and diag_3 == 'S':
                if include_wraps or ((row + 3*vertical_direction) >= 0 and (row + 3*vertical_direction) < n and (col + 3*horizontal_direction) >= 0 and (col + 3*horizontal_direction) < m):
                    if print_finds:
                        vertical_tags = ['', 'down', 'up']
                        horizontal_tags = ['', 'right', 'left']
                        print(f'{vertical_tags[vertical_direction]} {horizontal_tags[horizontal_direction]} diag at {row}, {col}')
                    xmas_count += 1
    return xmas_count
```

```python
xmas_total_count = 0

# Loop through every letter in the word search
for row in range(n):
    for col in range(m):

        # Are we at the start of a potential XMAS?
        if word_search[row][col] == 'X':
            # How many valid XMASs are there originating from this tile?
            xmas_total_count += count_XMAS(word_search,row,col, print_finds=True, include_wraps=False)
xmas_total_count
```

```python
# Part 2 wasn't wrap-arounds :(

# 4 different possible permutations:

# M.M
# .A.
# S.S

# M.S
# .A.
# M.S

# S.S
# .A.
# M.M

# S.M
# .A.
# S.M

# A is always in the centre, and M's and S's are either in the same col or row

word_search = input.split('\n')
word_search


# n = number of rows
# m = number of cols

n, m = len(word_search), len(word_search[0])

def is_xmas_cross(word_search,row,col):
    '''
    Input: word search and row/col that has an A that is not on the first/last row/col
    Output: True/False on whether there is a valid X-MAS cross around this index
    '''

    # Grab the letters that are diagonal
    up_left = word_search[row-1][col-1]
    up_right = word_search[row-1][col+1]
    down_left = word_search[row+1][col-1]
    down_right = word_search[row+1][col+1]

    # Is it a valid x-mas cross? See examples in comment above
    if [up_left, up_right, down_left, down_right] in [['M', 'M', 'S', 'S'], ['S', 'S', 'M', 'M'], ['M', 'S', 'M', 'S'], ['S', 'M', 'S', 'M']]:
        return True
    return False

```

```python
xmas_total_count = 0

# Loop through every letter in the word search
for row in range(n):
    for col in range(m):
        # Skip first and last row/col
        if row == 0 or col == 0 or row == n-1 or col == m-1:
            continue
        if word_search[row][col] == 'A':
            if is_xmas_cross(word_search,row,col):
                xmas_total_count += 1

xmas_total_count
```
